# **Theory Questions Answers**

Q-1 What is NumPy, and why is it widely used in Python?
- NumPy is a Python library for fast, efficient multidimensional array operations and numerical computing; it forms the backbone for most scientific, engineering, and data processing libraries in Python due to its speed, versatility, and powerful features for large datasets.

Q-2 How does broadcasting work in NumPy?
- Broadcasting allows NumPy to perform operations on arrays of different shapes by automatically expanding the smaller array along its missing dimensions so that arithmetic can proceed elementwise without explicit looping or copying.

Q-3 What is a Pandas DataFrame?
- A Pandas DataFrame is a two-dimensional, labeled data structure resembling a table, where data is organized in columns and rows; it supports diverse data types and is ideal for complex data analysis and manipulation tasks.

Q-4 Explain the use of the groupby() method in Pandas.
- The groupby() method in Pandas splits a DataFrame into groups based on column values, applies a function (such as aggregation, transformation, or filtering) to each group, and then combines the results into a new DataFrame; this is known as the "split-apply-combine" process.


Q-5 Why is Seaborn preferred for statistical visualizations?
- Seaborn is favored for statistical visualizations because it offers attractive default styles, integrates tightly with Pandas data structures, and provides built-in support for advanced plots (like regression, violin, and pair plots) with enhanced statistical summaries compared to Matplotlib alone.

Q-6 What are the differences between NumPy arrays and Python lists?
- NumPy arrays are fixed-type, homogeneous, and support vectorized operations, making them much faster and more memory-efficient for large or numerical data.

- Python lists are heterogeneous and inherently slower for mathematical operations, as they lack NumPy's optimizations.

Q-7 What is a heatmap, and when should it be used?
- A heatmap is a graphical representation of data where individual values are represented by color; it is used to visualize the magnitude and distribution of a matrix, such as correlation coefficients or confusion matrices, to help detect patterns, relationships, or anomalies quickly.

Q-8 What does the term “vectorized operation” mean in NumPy?
- A vectorized operation means performing computations on entire arrays or large blocks of data without explicit loops, resulting in faster and more efficient code due to NumPy's optimized C backend.


Q-9 How does Matplotlib differ from Plotly?
- Matplotlib is a foundational plotting library in Python, offering static, publication-quality plots with fine-grained customization.

- Plotly specializes in interactive, web-based plots and dashboards; it is easier for creating interactive visualizations, tooltips, and exporting to web formats.

Q-10 What is the significance of hierarchical indexing in Pandas?
- Hierarchical indexing (MultiIndex) in Pandas enables higher-dimensional data management within the two-dimensional DataFrame structure, supporting more complex data relationships, groupings, and flexible data selection.

Q-11 What is the role of Seaborn’s pairplot() function?
- The pairplot() function in Seaborn creates a matrix of scatterplots and histograms for each variable pair in a dataset, making it easy to visualize relationships, distributions, and possible trends among all variables at once.

Q-12 What is the purpose of the describe() function in Pandas?
- The describe() function provides quick summary statistics of a DataFrame, such as mean, median, standard deviation, min, max, and quartiles, giving a comprehensive snapshot of numerical columns for exploratory data analysis.



Q-13 Why is handling missing data important in Pandas?
- Handling missing data is crucial for ensuring accuracy, integrity, and reliability of analytical or statistical results, as unaddressed NaNs or nulls can distort calculations, conclusions, and visualizations.

Q-14 What are the benefits of using Plotly for data visualization?
- Plotly enables interactive, web-based plots with dynamic features like zoom, hover, and tooltips; it is user-friendly, works natively with Pandas and NumPy data, and supports building dashboards without extensive web development knowledge.

Q-15 How does NumPy handle multidimensional arrays?
- NumPy treats multidimensional arrays (ndarrays) as contiguous memory blocks, providing fast, vectorized access for any combination of dimensions using slicing, broadcasting, and shape manipulations.[14][11]

Q-16 What is the role of Bokeh in data visualization?
- Bokeh is a Python library designed for creating interactive, browser-based visualizations that can handle large or streaming datasets, integrating seamlessly with web applications and Jupyter notebooks.




Q-17 Explain the difference between apply() and map() in Pandas.
- map() is used for transforming values in a Pandas Series by substituting or mapping distinct values; it works elementwise.

- apply() can operate on either Series or DataFrames, applying functions across rows, columns, or individual elements; it is more general than map().

Q-18 What are some advanced features of NumPy?

- Advanced features include masked arrays, structured arrays, advanced indexing, broadcasting, linear algebra routines, random number generation, and interoperability with C and Fortran code.[14][13]

Q-19 How does Pandas simplify time series analysis?
- Pandas provides specialized functionality for date and time indexing, resampling, frequency conversion, rolling window statistics, and handling irregular time data, making time series alignment, transformation, and analysis much easier.

Q-20 What is the role of a pivot table in Pandas?
- A Pandas pivot table allows for dynamic data summarization by aggregating and rearranging data according to different categories, similar to Excel pivot tables, providing concise overviews and cross-tabulations.



Q-21 Why is NumPy’s array slicing faster than Python’s list slicing?

- NumPy's array slicing is faster because arrays are stored in contiguous blocks of memory, enabling efficient pointer-based operations, while Python lists are composed of references to objects, leading to slower access and modifications.[11][13][14]

Q-22 What are some common use cases for Seaborn?
- Common uses for Seaborn are statistical explorations (correlation heatmaps, regression plots), visualizing distributions (histograms, KDE plots), categorical comparisons (boxplots, barplots), and multi-plot matrices (pairplots, FacetGrid-based grids) for insightful data storytelling.

```python
# 2. Write a Pandas script to find the mean of a specific column in a DataFrame.

import pandas as pd

data = {"key1" : [1,2,3,4,5,6],
        "key2": [6,6,7,8,9,34],
        "key3" : [45,67,23,56,78,90]}

df = pd.DataFrame(data)
print(df["key3"].mean())
```

```python
# 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "A": [1, 2, 3, 4, 5],
    "B": [5, 4, 3, 2, 1],
    "C": [2, 3, 4, 5, 6]
}

df = pd.DataFrame(data)

corr = df.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.show()
```

```python
# 7. Write a program to perform element-wise multiplication of two NumPy arrays.
import numpy as np

arr1 = np.array([[2, 2, 9], [3, 4, 9], [2, 7, 2]])
arr2 = np.array([[9, 9, 5], [5, 3, 7], [9, 6, 1]])

result = arr1 * arr2
print(result)
```

```python
# 8. Create a line plot with multiple lines using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(1, 10, 100)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) + np.cos(x)

plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.plot(x, y3, label='sin(x) + cos(x)')

plt.title('Multiple Line Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

plt.show()
```
